```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore") 
%matplotlib inline
```

### 0. Importing Data

```python
#데이터 불러오기
df = pd.read_csv('Bitcoin Historical Data - Investing.com.csv')
#데이터 복사
origin_data = df.copy()
df.head()
```

```python
# 데이터 정렬 변경 - 새로운 인덱스 생성
new_index = range(1128)
new_index = sorted(new_index, reverse=True)
print(list(new_index))
```

```python
# 새로운 인덱스 부여
df['new_index'] = new_index
df = df.sort_values('new_index',ascending=True)
df.set_index('new_index', drop=True, inplace=True)
```

```python
df.shape
```

### 2. 파생변수 생성

```python
df['Change %'] = df['Change %'].str.replace("%","")
```

```python
df['predict'] = np.nan
```

```python
df.head()
```

```python
#종속변수  predict ---- 일주일 뒤의 가격의 등락 여부(0,1)

date = len(df['Date'])-7 #마지막 주는 7일 뒤 데이터가 없기 때문에 종속변수에서 제외

for i in range(date) :
  df.loc[i,'predict'] = np.where(float(df.loc[i+7,'Change %']) > 0,0,1)
```
