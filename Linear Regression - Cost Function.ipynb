### 선형회귀 - 최소제곱법 구현

```python
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
```

```python
mx = np.mean(x)
my = np.mean(y)
```

```python
divisor = sum([(i-mx)**2 for i in x])
```

```python
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i]-mx)*(y[i]-my)
  return d
divided = top(x, mx, y, my)
```

```python
a = divided/divisor
```

```python
b = my - (mx*a)
```

```python
print("기울기 a =", a)
print("y 절편 b =", b)
```

### 선형회귀 - 평균제곱오차(Mean Square Error)

```python
import numpy as np

#가상의 기울기 a와 y 절편 b
temp_a_b=[3,76]

# x 값과 y값
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# y=ax + b에 a,b 값 대입하여 결과를 출력하는 함수
def predict(x):
   return temp_a_b[0]*x + temp_a_b[1]

# MSE 함수
def mse(y, y_hat):
   return ((y - y_hat) ** 2).mean()

# MSE 함수를 각 y값에 대입하여 최종 값을 구하는 함수
def mse_val(predict_result,y):
   return mse(np.array(predict_result), np.array(y))

# 예측값이 들어갈 빈 리스트
predict_result = []

# 모든 x값을 한 번씩 대입하여 predict_result 리스트완성.
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))

# 최종 MSE 출력
print("MSE 최종값: " + str(mse_val(predict_result,y)))
```
